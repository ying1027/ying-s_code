## Age Demographics

* Categorize players using the age groups. Hint: use pd.cut()
  * each group is for 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* round the percentage column to two decimal points
* Display Age Demographics Table

```python
age=purchase_data.loc[:,['Age','SN']]
d=age.drop_duplicates(subset='SN')
bin=[0,9,14,19,24,29,34,39,1000]
d['bins']=pd.cut(x=age['Age'],bins=bin,labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
d1=d.groupby('bins').count()
d1=d1.drop(columns=['SN'])
d1.columns=['Total Count']
d1['Percentage of Players']=round((d1['Total Count']/d1['Total Count'].sum())*100,2)
d1=d1.reindex(columns =["Percentage of Players",'Total Count' ])
d1
```

## Purchasing Analysis (Age)

* The below each broken into groups of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
  * Average Purchase Total per Person by Age Group
* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Display the summary data frame

```python
pur=purchase_data.loc[:,['Age','Price']]


bin=[0,9,14,19,24,29,34,39,1000]
pur['bins']=pd.cut(x=pur['Age'],bins=bin,labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
e=pur.groupby('bins').count()
e=e.drop(columns=['Price'])
pur_=pur.drop(columns=['Age'])
pur_=pur_.groupby('bins').aggregate([sum,pd.DataFrame.mean])

age=purchase_data.loc[:,['Age','SN']]
d=age.drop_duplicates(subset='SN')
bin=[0,9,14,19,24,29,34,39,1000]
d['bins']=pd.cut(x=age['Age'],bins=bin,labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
d1=d.groupby('bins').count()

pieces2=[e,pur_]
summary_data=pd.concat(pieces2,axis=1)
summary_data['Average Purchase Total per Person']=summary_data[('Price', 'sum')]/d1['Age']


summary_data.columns=['Purchase Count','Total Purchase Value','Average Purchase Price','Average Purchase Total per Person']

summary_data
summary_data.round(2)
```
